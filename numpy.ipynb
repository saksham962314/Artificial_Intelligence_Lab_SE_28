```python
import numpy as np

# Create two arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Addition of arrays
sum_array = a + b

# Multiplication of arrays
mul_array = a * b

print("Array A:", a)
print("Array B:", b)
print("Addition:", sum_array)
print("Multiplication:", mul_array)
```
