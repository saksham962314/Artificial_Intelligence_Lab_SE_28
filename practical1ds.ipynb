PS: WAP to compute factorial of a given number
**Prompt: WAP to compute factorial of a given number take input from user**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

while True:
    try:
        num_str = input("Enter a non-negative integer: ")
        num = int(num_str)
        result = factorial(num)
        print(f"The factorial of {num} is {result}")
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

ps:WAP to identify the given number is odd or even
**prompt:WAP to identify the given number is odd or even take input from user**

```python
while True:
    try:
        num_str = input("Enter an integer to check if it's odd or even: ")
        num = int(num_str)

        if num % 2 == 0:
            print(f"The number {num} is Even.")
        else:
            print(f"The number {num} is Odd.")
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

ps:WAP to identify the entered number is palindrome or not.
**prompt:WAP to identify the entered number is palindrome or not.take input from user**

```python
while True:
    try:
        num_str = input("Enter a number to check if it's a palindrome: ")
        # Convert to string to easily check for palindrome property
        # A number is a palindrome if it reads the same forwards and backwards.
        if num_str == num_str[::-1]:
            print(f"The number {num_str} is a palindrome.")
        else:
            print(f"The number {num_str} is not a palindrome.")
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

ps:Develop an Expert System that provides simple decision-making.
**prompt:Develop an Expert System that provides simple decision-making.take input from user**


```python
def get_expert_system_rules():
    """
    Defines the rules for a simple expert system.
    Each rule consists of a condition and a recommendation.
    """
    rules = [
        {
            "condition": {"weather": "sunny"},
            "recommendation": "Go for a walk or a picnic in the park."
        },
        {
            "condition": {"weather": "rainy"},
            "recommendation": "Stay indoors and read a book or watch a movie."
        },
        {
            "condition": {"weather": "cloudy"},
            "recommendation": "Visit a museum or a coffee shop."
        },
        {
            "condition": {"temperature": "cold"},
            "recommendation": "Wear warm clothes and have a hot beverage."
        },
        {
            "condition": {"temperature": "hot"},
            "recommendation": "Drink plenty of water and find shade."
        }
    ]
    return rules

# Display the defined rules
expert_system_rules = get_expert_system_rules()
print("Expert System Rules:")
for i, rule in enumerate(expert_system_rules):
    print(f"Rule {i+1}: IF {rule['condition']} THEN {rule['recommendation']}")
```

```python
current_weather = input("Enter the current weather condition (e.g., sunny, rainy, cloudy): ").lower()
current_temperature = input("Enter the current temperature condition (e.g., cold, hot): ").lower()

print(f"\nUser Input Collected:\nWeather: {current_weather}\nTemperature: {current_temperature}")
```

```python
applicable_recommendations = []

for rule in expert_system_rules:
    condition = rule['condition']

    # Check for weather condition match
    if 'weather' in condition and condition['weather'] == current_weather:
        applicable_recommendations.append(rule['recommendation'])

    # Check for temperature condition match
    if 'temperature' in condition and condition['temperature'] == current_temperature:
        applicable_recommendations.append(rule['recommendation'])


# Display recommendations
print("\n--- Expert System Recommendation ---")
if not applicable_recommendations:
    print("No specific recommendations can be made based on your input.")
else:
    print("Based on your input, here are some recommendations:")
    for rec in applicable_recommendations:
        print(f"- {rec}")
```
